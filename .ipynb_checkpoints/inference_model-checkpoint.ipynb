```python
def split_and_average(value):
    if isinstance(value, str):
        float_list = list(map(float, value.split('|'))) 
        return sum(float_list) / len(float_list) 
    elif pd.isnull(value):
        return value
    else:
        return value 
    
def filter_rows(value):
    if isinstance(value, str):  # Check if the value is a string
        num_list = value.split('|')
        # Only keep the row if the num_list has exactly 1 entry
        return len(num_list) == 1
    return True
```

```python
def train_linear_regression(X, y):
    model = LinearRegression()
    grid = {"fit_intercept": [True, False], "positive": [True, False]}
    kf = KFold(n_splits=10, shuffle=True, random_state=1)
    grid_search = GridSearchCV(model, grid, cv=kf)
    grid_search.fit(X, y)
    return grid_search.best_estimator_, grid_search.best_params_


def train_logistic_regression(X, y):
    model = LogisticRegression(solver='liblinear')
    grid = {"C": np.logspace(-3, 3, 7), "penalty": ["l1", "l2"]}
    kf = KFold(n_splits=10, shuffle=True, random_state=1)
    grid_search = GridSearchCV(model, grid, cv=kf)
    grid_search.fit(X, y)
    return grid_search.best_estimator_, grid_search.best_params_


# Function to calculate odds ratios from model coefficients
def calculate_odds_ratios(model):
    coefficients = model.coef_[0]
    odds_ratios = np.exp(coefficients)
    return coefficients, odds_ratios


# Function to calculate covariance matrix and standard errors for logistic regression
def calculate_standard_errors(X, y, model):
    X_with_const = np.hstack([np.ones((X.shape[0], 1)), X])  # Adding intercept manually
    pred_probs = model.predict_proba(X)[:, 1]
    V = np.diagflat(pred_probs * (1 - pred_probs))  # Diagonal matrix with variances
    cov_matrix = np.linalg.inv(np.dot(X_with_const.T, np.dot(V, X_with_const)))  # Covariance matrix
    standard_errors = np.sqrt(np.diag(cov_matrix)[1:])  # Exclude the intercept term
    return standard_errors


# Function to calculate confidence intervals for logistic regression
def calculate_confidence_intervals(coefficients, standard_errors, z=1.96):
    ci_lower = coefficients - z * standard_errors
    ci_upper = coefficients + z * standard_errors
    ci_lower_exp = np.exp(ci_lower)
    ci_upper_exp = np.exp(ci_upper)
    return ci_lower_exp, ci_upper_exp


# Function to calculate standard errors and confidence intervals for the coefficients of linear regression
def calculate_confidence_intervals_lr(X, y_true, model, alpha=0.05):
    # Calculate residuals
    predictions = model.predict(X)
    residuals = y_true - predictions

    # Get the degrees of freedom
    n, p = X.shape
    degrees_of_freedom = n - p - 1

    # Variance of residuals (mean squared error of residuals)
    residual_variance = np.sum(residuals**2) / degrees_of_freedom

    # Covariance matrix of the coefficients
    X_with_const = np.hstack([np.ones((X.shape[0], 1)), X])  # Add intercept manually if necessary
    cov_matrix = np.linalg.inv(X_with_const.T @ X_with_const) * residual_variance

    # Standard errors of the coefficients
    standard_errors = np.sqrt(np.diag(cov_matrix)[1:])  # Exclude the intercept

    # Calculate the z-score for the given confidence level
    z = norm.ppf(1 - alpha / 2)

    # Calculate the confidence intervals
    coefficients = model.coef_
    ci_lower = coefficients - z * standard_errors
    ci_upper = coefficients + z * standard_errors

    return standard_errors, ci_lower, ci_upper




# Function to print model performance and results
def print_results(X, y, model, coefficients, odds_ratios, ci_lower, ci_upper):
    predictions = model.predict(X)
    accuracy = accuracy_score(y, predictions)
    print(f"Accuracy: {accuracy}\n")
    print("Classification Report:")
    print(classification_report(y, predictions))
    print("\nCoefficients, Odds Ratios, and 95% Confidence Intervals:")
    
    feature_names = X.columns
    for feature, coef, or_, ci_l, ci_u in zip(feature_names, coefficients, odds_ratios, ci_lower, ci_upper):
        print(f"{feature}: Coefficient={coef:.4f}, OR={or_:.4f}, 95% CI=[{ci_l:.4f}, {ci_u:.4f}]")


# Main function to run the workflow
def run_workflow(X, y):
    # Train the model using GridSearchCV
    best_model, best_params = train_logistic_regression(X, y)
    print(f"Tuned hyperparameters (best parameters): {best_params}\n")

    # Calculate odds ratios and coefficients
    coefficients, odds_ratios = calculate_odds_ratios(best_model)

    # Calculate standard errors and confidence intervals
    standard_errors = calculate_standard_errors(X, y, best_model)
    ci_lower, ci_upper = calculate_confidence_intervals(coefficients, standard_errors)

    # Print results
    print_results(X, y, best_model, coefficients, odds_ratios, ci_lower, ci_upper)
```

```python
# Function to filter rows based on the value at a specific index in the one-hot encoded column
def filter_by_one_hot(df, col_name, index):
    return df[df[col_name].apply(lambda x: x[index] == 1)]

filtered_df = filter_by_one_hot(all_data, 'Element_abbr_encoded', 0)


# Function to expand the one-hot encoded column into multiple binary columns
def expand_one_hot(df, col_name):
    # Get the length of the longest vector
    max_len = max(df[col_name].apply(len))
    
    # Create new binary columns for each index in the one-hot encoded vector
    for i in range(max_len):
        df[f'one_hot_{i}'] = df[col_name].apply(lambda x: x[i] if i < len(x) else 0)
    
    # Optionally, drop the original one-hot column
    #df.drop(columns=[col_name], inplace=True)
    
    return df
```
